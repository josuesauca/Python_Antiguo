```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

contamina = pd.read_csv("C:\Segundo Ciclo\Teoria de la Distribucion y Probabilidad\Recuperacion Prueba\8) ASOCIACIONES LINEALES Y LA CORRELACION\Contaminacion Atmosferica.csv")
contamina.info()
contamina.head()
```

```python

#No estan correlacionada o sin relacion

plt.scatter(contamina['Velocidad_viento'],contamina['Dias_Lluvia'])
plt.title('Dias Lluvia y Velocidad Viento')
plt.xlabel('Velocidad viento')
plt.ylabel('Dias Lluvia')

plt.show()
```

```python

#Correlacion Negativa

"""
mayor es la temperatura menor es la contaminacion
"""

plt.scatter(contamina['Temperatura'],contamina['Contaminacion_SO2'])
plt.title('Contaminacion SO2 y Temperatura')
plt.xlabel("Temperatura")
plt.ylabel("Contaminacion SO2")
plt.show()
```

```python

#Correlacion positiva

plt.scatter(contamina['Habitantes'],contamina['Fabricas'])
plt.title('Contaminacion SO2 y Temperatura')
plt.xlabel("Habitantes")
plt.ylabel("Fabricas")
plt.show()
```

```python

#Todas las graficas

sns.pairplot(contamina)
plt.show()
```
